## Need to remove Textbox1, Textbox4, Textbox6, Textbox7, and Textbox8 columns. They are grand totals for the entire data set but not needed. Also need to rename 'AreaType' to 'County'

```python
#testing on 2013 data.
harvest_2013 = harvest_2013.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2013 = harvest_2013.rename(columns={'AreaType': 'County'})
```

```python
harvest_2013.head()
```

```python
#repeating process above for rest of harvest files.
harvest_2014 = harvest_2014.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2014 = harvest_2014.rename(columns={'AreaType': 'County'})
harvest_2015 = harvest_2015.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2015 = harvest_2015.rename(columns={'AreaType': 'County'})
harvest_2016 = harvest_2016.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2016 = harvest_2016.rename(columns={'AreaType': 'County'})
harvest_2017 = harvest_2017.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2017 = harvest_2017.rename(columns={'AreaType': 'County'})
harvest_2018 = harvest_2018.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2018 = harvest_2018.rename(columns={'AreaType': 'County'})
harvest_2019 = harvest_2019.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2019 = harvest_2019.rename(columns={'AreaType': 'County'})
harvest_2020 = harvest_2020.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2020 = harvest_2020.rename(columns={'AreaType': 'County'})
harvest_2021 = harvest_2021.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2021 = harvest_2021.rename(columns={'AreaType': 'County'})
harvest_2022 = harvest_2022.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2022 = harvest_2022.rename(columns={'AreaType': 'County'})
harvest_2023 = harvest_2023.drop(columns=['Textbox1', 'Textbox4', 'Textbox6', 'Textbox7', 'Textbox8'])
harvest_2023 = harvest_2023.rename(columns={'AreaType': 'County'})
```

```python
harvest_2023.head()
```

```python
#next need to add a 'season' column with the corresponding file year.
harvest_2013['Season'] = '2013'
harvest_2013.head()
```

```python
#repeat the process for the rest of the files.
harvest_2014['Season'] = '2014'
harvest_2015['Season'] = '2015'
harvest_2016['Season'] = '2016'
harvest_2017['Season'] = '2017'
harvest_2018['Season'] = '2018'
harvest_2019['Season'] = '2019'
harvest_2020['Season'] = '2020'
harvest_2021['Season'] = '2021'
harvest_2022['Season'] = '2022'
harvest_2023['Season'] = '2023'
harvest_2023.head()
```

## Need to now merge the files to one file.

```python
deer_harvest = pd.concat([harvest_2013, harvest_2014, harvest_2015, harvest_2016, harvest_2017, harvest_2018, harvest_2019, harvest_2020, harvest_2021, harvest_2022, harvest_2023])
```

```python
deer_harvest.info()
```

```python
deer_harvest.head()
```

```python
#removed commas to convert dtypes to int.
deer_harvest = deer_harvest.replace(',','', regex=True)
```

```python
deer_harvest['AntlerMaleCnt'] = deer_harvest['AntlerMaleCnt'].astype(int)
deer_harvest['FemaleCnt'] = deer_harvest['FemaleCnt'].astype(int)
deer_harvest['TotalHarvestCount'] = deer_harvest['TotalHarvestCount'].astype(int)
```
